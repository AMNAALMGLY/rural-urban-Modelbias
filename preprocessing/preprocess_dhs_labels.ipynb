```python
#filter data that are only on nature paper 
dhs_sustain=pd.read_csv('/network/scratch/a/amna.elmustafa/rural-urban-Modelbias/batchers/dhs_final_labels.csv')
#dhs_sustain=dhs_sustain.interpolate(method='polynomial',order=3)

old=pd.read_csv('/network/scratch/a/amna.elmustafa/rural-urban-Modelbias/data/dhs_clusters.csv')
#preprocessing 
#dhs_sustain.loc[:,dhs_labels]=dhs_sustain.loc[:,dhs_labels].apply(lambda x : (x-x.mean())/x.std())

dhs_sustain[['lat', 'lon']] = dhs_sustain[['lat', 'lon']].apply(lambda x: x.astype(np.float32))
old[['lat', 'lon']] = old[['lat', 'lon']].apply(lambda x: x.astype(np.float32))

lons=old['lon'].to_numpy().astype(np.float32)
lats=old['lat'].to_numpy().astype(np.float32)
match =dhs_sustain[(dhs_sustain['lon'].isin(list(lons))) & (dhs_sustain['lat'].isin(list(lats)))].reset_index(drop=True)
```

```python
out=old.merge(match, left_on=['lat','lon'], right_on=['lat','lon']).reindex(columns=list(old.columns)+dhs_labels)
out['year']=old['year']
out['women_bmi'].quantile(0.99)
#out.to_csv('./match_unprocessed.csv')
```

```python
#filling missing values by knn imputer after scaling values

from sklearn.impute import KNNImputer
#scaling:
match.loc[:,dhs_labels]=match.loc[:,dhs_labels].apply(lambda x : (x-x.mean())/(x.std()))
#imputing
imputer = KNNImputer(n_neighbors=5)
df = pd.DataFrame(imputer.fit_transform(match[dhs_labels]),columns = dhs_labels,dtype=np.float32)
match[dhs_labels]=df[dhs_labels]
```

```python
#filling missing values with knn , but after minQuantile scaler 
from sklearn.impute import KNNImputer
#scaling:
match.loc[:,dhs_labels]=match.loc[:,dhs_labels].apply(lambda x : (x-x.min())/(x.quantile(0.99)-x.min()))
#imputing
print(match)
imputer = KNNImputer(n_neighbors=5)
df = pd.DataFrame(imputer.fit_transform(match[dhs_labels]),columns = dhs_labels,dtype=np.float32)
match[dhs_labels]=df[dhs_labels]
#Save this method of preprocessing
match[['lat', 'lon']] = match[['lat', 'lon']].apply(lambda x: x.astype(np.float32))

out=old.merge(match, left_on=['lat','lon'], right_on=['lat','lon']).reindex(columns=list(old.columns)+dhs_labels)
out['year']=old['year']
                              
#out.to_csv('./dhs_labels_quantile.csv')                                                      
```

```python
#after scaling and imputing , do some transformation (sqrt)
match.loc[:,dhs_labels]=match.loc[:,dhs_labels].apply(lambda x : np.sqrt(x))
#
#Save this method of preprocessing
match[['lat', 'lon']] = match[['lat', 'lon']].apply(lambda x: x.astype(np.float32))

out=old.merge(match, left_on=['lat','lon'], right_on=['lat','lon']).reindex(columns=list(old.columns)+dhs_labels)
out['year']=old['year']
                              
out.to_csv('./dhs_labels_quantile_sqrt.csv')                                                      
```
